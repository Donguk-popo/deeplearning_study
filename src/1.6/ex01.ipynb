```python
from pandas import read_csv

column_names = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE']

boston_pd = read_csv("../../data/1.6_data/housing.csv",header=None, names=column_names)
boston_pd.head()
```

```python
boston_pd.drop(index=0,inplace=True)
```

```python
boston_pd.info()
```

```python
boston_pd["PRICE"] = boston_pd["PRICE"].astype(float)
boston_pd["RM"] = boston_pd["RM"].astype(float)
boston_pd["LSTAT"] = boston_pd["LSTAT"].astype(float)
```

```python
import plotly.express as px

fig = px.histogram(boston_pd, x = 'PRICE')
fig.show()
```

```python
from sklearn.model_selection import train_test_split

X = boston_pd.drop('PRICE', axis=1)
y = boston_pd['PRICE']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error               # 사이킷런에서 제공하는 함수 평균제곱오차를 계산합니다

pred_tr = reg.predict(X_train)                               # reg : 훈련된 회귀 모델 객체
pred_test = reg.predict(X_test)                              # predict : 훈련 데이터 (X_test) 에 대한 모델의 예측 값을 계산합니다.
rmse_tr = (np.sqrt(mean_squared_error(y_train, pred_tr)))    # np.sqrt : 평균 제곱 오차의 제곱근을 계산하여 RMSE를 구한다.
rmse_test = (np.sqrt(mean_squared_error(y_test, pred_test)))

print(rmse_tr)
print(rmse_test)
```
